Libs necessárias

```python
%matplotlib inline
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import auc, roc_curve
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.model_selection import StratifiedKFold, KFold, GridSearchCV, train_test_split
from sklearn import datasets, linear_model, metrics, ensemble, svm, neighbors, naive_bayes, tree, ensemble
import xgboost

plt.style.use('ggplot')
warnings.filterwarnings("ignore")
```

```python
X_boston, y_boston = datasets.load_boston(return_X_y=True)

X_train_boston, X_test_boston, y_train_boston, y_test_boston = train_test_split(X_boston, y_boston, test_size=0.25,random_state=42)

stdzer = StandardScaler().fit(X_train_boston)
X_train_boston_std = stdzer.transform(X_train_boston)
X_test_boston_std = stdzer.transform(X_test_boston)
```

# Criação e comparação de modelos

###### Tuning hyperparameters via GridSearch

```python
model_rf = ensemble.RandomForestRegressor(random_state=42)

n_estimators = [50, 100, 200]
max_depth = [3, 5, 7]

param_grid = {'n_estimators': n_estimators, 
              'max_depth': max_depth}

gs = GridSearchCV(estimator=model_rf, param_grid=param_grid, scoring='neg_mean_squared_error', cv=10, n_jobs=-1)
gs = gs.fit(X_train_boston_std, y_train_boston)

print(gs.best_params_)
```

```python
model_gb = ensemble.GradientBoostingRegressor(random_state=42)
n_estimators = [50, 100, 200]
max_depth = [3, 5, 7]
learning_rate = [ 0.1, 0.05, 0.01]

param_grid = {'n_estimators': n_estimators, 
              'max_depth': max_depth, 
              'learning_rate': learning_rate}

gs = GridSearchCV(estimator=model_gb, param_grid=param_grid, scoring='neg_mean_squared_error', cv=10, n_jobs=-1)
gs = gs.fit(X_train_boston_std, y_train_boston)

print(gs.best_params_)
```

```python
model_xgb = xgboost.XGBRegressor(random_state=42)
n_estimators = [50, 100, 200]
max_depth = [3, 5, 7]
learning_rate = [ 0.1, 0.05, 0.01]

param_grid = {'n_estimators': n_estimators, 
              'max_depth': max_depth, 
              'learning_rate': learning_rate}

gs = GridSearchCV(estimator=model_xgb, param_grid=param_grid, scoring='neg_mean_squared_error', cv=10, n_jobs=-1)
gs = gs.fit(X_train_boston_std, y_train_boston)

print(gs.best_params_)
```

## Regressão

```python
r_r = {}
#r_r['SGD'] = {'model': linear_model.SGDRegressor(max_iter=1000)}
r_r['SVM'] = {'model': svm.SVR()}
r_r['KNN'] = {'model': neighbors.KNeighborsRegressor(n_neighbors=5, weights='distance')}
r_r['DT'] = {'model': tree.DecisionTreeRegressor(random_state=42)}
r_r['RF'] = {'model': ensemble.RandomForestRegressor(max_depth= 7, n_estimators=200, random_state=42)}
r_r['GB'] = {'model': ensemble.GradientBoostingRegressor(learning_rate=0.05, max_depth=3, n_estimators=200, random_state=42)}
r_r['XGB'] = {'model': xgboost.XGBRegressor(learning_rate=0.1, max_depth=5, n_estimators=100, random_state=42)}
```

```python
std_zer = StandardScaler()
kf = KFold(n_splits=5, random_state=42, shuffle=True)

for key in r_r:
    print('processando {}...'.format(key))
    rmse = []
    for idx_train, idx_test in kf.split(X_boston, y_boston):
        std_zer.fit(X_boston[idx_train])
        X_boston_train_std = std_zer.transform(X_boston[idx_train])
        X_boston_test_std = (X_boston[idx_test])
        y_train, y_test = y_boston[idx_train] ,y_boston[idx_test]
        
        r_r[key]['model'].fit(X_boston_train_std, y_train)
        y_pred = r_r[key]['model'].predict(X_boston_test_std)
        rmse.append( np.sqrt( metrics.mean_squared_error(y_test, y_pred)))
    r_r[key]['rmse'] = np.mean(rmse)
```
